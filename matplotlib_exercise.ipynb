## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('data/hist_data.csv')
plt.hist(df['aデータ'], alpha=0.5, bins=30, label='aデータ', color='orange')
plt.hist(df['bデータ'], alpha=0.4, bins=30, label='bデータ', color=(0, 1, 0.7))

plt.xlabel('温度(℃)')
plt.ylabel('度数')

plt.rcParams.update({'font.size': 8})
plt.gcf().set_size_inches(4.5, 3)
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv('data/scatter_data.csv')

plt.scatter(df['length (cm)'][df['class'] == 0], df['width (cm)'][df['class'] == 0], c='red', label='class0')

plt.scatter(df['length (cm)'][df['class'] == 1], df['width (cm)'][df['class'] == 1], c='blue', label='class1')

plt.scatter(df['length (cm)'][df['class'] == 2], df['width (cm)'][df['class'] == 2], c='green', label='class2')

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.rcParams.update({'font.size': 8})
plt.gcf().set_size_inches(4.5, 3)
plt.legend()
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.grid()

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
#折れ線
df = pd.read_csv('data/line_data.csv')
plt.plot(df["期間"], df["件数"], marker='o')

plt.xlabel('期間')
plt.ylabel('件数')

plt.rcParams.update({'font.size': 8})
plt.gcf().set_size_inches(4.5, 3)
plt.title('期間ごとの件数の推移')
plt.grid()
plt.xticks(rotation=30)
plt.show()

#棒
plt.bar(df["期間"], df["費用"])

plt.xlabel('期間')
plt.ylabel('費用(円)')

plt.rcParams.update({'font.size': 8})
plt.gcf().set_size_inches(4.5, 3)
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=30)
plt.grid()
plt.show()
```
